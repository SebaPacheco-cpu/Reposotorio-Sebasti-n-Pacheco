1. Considerando las variables earn, height y age, vamos a suponer que nuestro dataset
fuera una población completa. (Prepara los datos adecuadamente).
a. Realiza 5 pruebas con muestras de tamaño n = 25, en cada caso, con niveles
de significancia diferentes pero inferiores a 0,1, para confirmar o rechazar las
siguientes hipótesis considerando las alternativas. Crea para ello una fórmula
que reciba los parámetros adecuados y responda "con nivel de confianza del
...%, se acepta la hipótesis nula/alternativa".

```python
import pandas as pd
from scipy import stats
def test_hipotesis(earning_cleaned , columna, valor_hipotesis, alternativa): #Definimos la función
    # Realizar la prueba t de una muestra
    t_stat, p_value = stats.ttest_1samp(earning_cleaned [columna], popmean=valor_hipotesis)

    # Determinar si se acepta o rechaza la hipótesis nula
    if alternativa == "!=":
        if p_value < 0.1:
            decision = "rechazar la hipótesis nula"
        else:
            decision = "aceptar la hipótesis nula"
    elif alternativa == ">":
        if p_value < 0.1:
            decision = "rechazar la hipótesis nula"
        else:
            decision = "aceptar la hipótesis nula"
    elif alternativa == "<":
        if p_value < 0.1:
            decision = "rechazar la hipótesis nula"
        else:
            decision = "aceptar la hipótesis nula"

    # Mostrar el resultado de la prueba
    print(f"Para la variable '{columna}':")
    print(f"Hipótesis nula (H0): x = {valor_hipotesis}")
    print(f"Hipótesis alternativa (H1): x {alternativa} {valor_hipotesis}")
    print(f"P-valor: {p_value}")
    print(f"Con un nivel de significancia de 0.1, se decide {decision}\n")

# Realizar pruebas de hipótesis para las variables especificadas
test_hipotesis(earning_cleaned, 'earn', 20000, '!=')
test_hipotesis(earning_cleaned, 'height', 55, '!=')
test_hipotesis(earning_cleaned, 'age', 44, '!=')
test_hipotesis(earning_cleaned, 'earn', 35000, '!=')
test_hipotesis(earning_cleaned, 'height', 68, '!=')
test_hipotesis(earning_cleaned, 'age', 56, '!=')
test_hipotesis(earning_cleaned, 'earn', 25000, '>')
test_hipotesis(earning_cleaned, 'age', 63, '>')
test_hipotesis(earning_cleaned, 'age', 51, '>')
test_hipotesis(earning_cleaned, 'earn', 32000, '<')
test_hipotesis(earning_cleaned, 'age', 70, '<')
test_hipotesis(earning_cleaned, 'age', 62, '<')
```

2. Crea una función que estime la proporción de hombres en el dataset, considerando
una muestra de 50 individuos, con niveles de significancia de 0,05 y 0,01. Escribe con
palabras tus resultados ejecutar la función e interpreta. Compara con la proporción
real.

```python

def estimar_proporcion_hombres(data , muestra_size, nivel_significancia=0.05):
    # Tomamos una muestra aleatoria de tamaño muestra_size
    muestra = earning_cleaned.sample(n=muestra_size, random_state=42)
    
    # Calcular la proporción de hombres en la muestra
    prop_hombres_muestra = muestra['male'].mean()
    
    # Calcular la proporción real de hombres en el dataset (earning_cleaned) completo
    prop_hombres_real = earning_cleaned['male'].mean()
    
    # Realizar una prueba de hipótesis para comparar la proporción de hombres en la muestra con la proporción real
    z_stat, p_value = stats.ttest_1samp(muestra['male'], prop_hombres_real)
    
    # Imprimir los resultados de la prueba de hipótesis
    print(f"Proporción de hombres en la muestra: {prop_hombres_muestra}")
    print(f"Proporción real de hombres en el dataset: {prop_hombres_real}")
    print(f"P-valor: {p_value}")
    
    # Comparar el p-valor con el nivel de significancia dado
    if p_value < nivel_significancia:
        print(f"Con un nivel de significancia de {nivel_significancia}, se rechaza la hipótesis nula.")
    else:
        print(f"Con un nivel de significancia de {nivel_significancia}, no se rechaza la hipótesis nula.")

# función con un nivel de significancia de 0.05
print("Resultados con nivel de significancia de 0.05:")
estimar_proporcion_hombres(earning_cleaned, muestra_size=50, nivel_significancia=0.05)

#  de significancia de 0.01
print("\nResultados con nivel de significancia de 0.01:")
estimar_proporcion_hombres(earning_cleaned, muestra_size=50, nivel_significancia=0.01)
```

La proporción en la muestra es del 38%, mientras que la proporción real de hombres en el conjunto de datos completos es aprox. del 37.78%. Por lo tanto, en ambos casos, con niveles de significancia de 0.05 y 0.01 respectivamente, no se rechaza la hipótesis nula, lo que significa que no hay evidencia sufiencte para afirmar que la proporción de hombres difiere significativamente de la proporción real en el conjunto de datos completos.
